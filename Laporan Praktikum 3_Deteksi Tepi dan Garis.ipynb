```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import skimage
```

```python
image = cv2.imread('2 (1).jpg')
```

```python
cv2.imshow('Gambar parkir', image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
gray = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
edges = cv2.Canny(image, 100, 150)
```

```python
cv2.imshow('Gambar parkir', edges)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
fig, axs = plt.subplots(1,2, figsize=(10,10))
ax = axs.ravel()

ax[0].imshow(gray, cmap = 'gray')
ax[0].set_title("Gambar asli")

ax[1].imshow(edges, cmap = 'gray')
ax[1].set_title("Gambar yang udah dideteksi")
```

```python
lines = cv2.HoughLinesP(edges, 1, np.pi/180, 50, maxLineGap = 7)
image_line = image.copy()
```

```python
for line in lines:
    x1, y1, x2, y2 = line[0]
    cv2.line(image_line, (x1, y1), (x2, y2), (100, 8, 255), 3)
```

```python
fig, axs = plt.subplots(1,2, figsize=(10,10))
ax = axs.ravel()

ax[0].imshow(gray, cmap = 'gray')
ax[0].set_title("Gambar asli")

ax[1].imshow(image_line, cmap = 'gray')
ax[1].set_title("Gambar yang udah dideteksi")
```
